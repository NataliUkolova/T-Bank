```python
''' Для полученной структуры напишите код, который выполняет следующие действия (4 балла)

Добавление любимых фраз Копатыча и Лосяша
Поиск серий, в которых участвуют оба персонажа
Удаление серии Герой Плутона из списка серий Лосяша
Замена цвета Копатыча на Светло-коричневый
Добавление данных о любом другом смешарике пользователем
Подсчет количества записанных смешариков
Создание второй структуры, идентичной первой, удаление оттуда записи
про Копатыча так чтобы оригинальная структура не поменялась
'''

smesharikis = { 
    "Лосяш": {"Цвет": "Терракотовый", 
    "Серии": ["Герой Плутона", "Бабочка", "Долгая рыбалка"], "Любимая фраза": []}, 
    "Копатыч": {"Цвет": "Коричневый", "Серии": ["Долгая рыбалка", "Танцор Диско", "Кулинария" ],
                "Любимая фраза": []}
}

#Добавление любимых фраз Копатыча и Лосяша
smesharikis["Лосяш"]["Любимая фраза"].append(input('Введите любимую фразу для Лосяша: '))
smesharikis["Копатыч"]["Любимая фраза"].append(input('Введите любимую фразу для Копатыча: '))
print(smesharikis)

#Поиск серий, в которых участвуют оба персонажа
print('Общие серии: ', *set(smesharikis["Лосяш"]["Серии"]).intersection(set(smesharikis["Копатыч"]["Серии"])))

#Удаление серии Герой Плутона из списка серий Лосяша
smesharikis["Лосяш"]["Серии"].remove("Герой Плутона")
print(smesharikis)

#Замена цвета Копатыча на Светло-коричневый
smesharikis["Копатыч"]["Цвет"] = "Светло-коричневый"
print(smesharikis)

#Добавление данных о любом другом смешарике пользователем
print('Сейчас есть информация о ', *smesharikis)
name = input('Введите имя нового смешарика ')
smesharikis[name] = dict()
smesharikis[name]["Цвет"] = input('Введите цвет ')
smesharikis[name]["Серии"] = []
smesharikis[name]["Серии"].append(input('Введите название серии '))
print(smesharikis)

#Подсчет количества записанных смешариков
print(len(smesharikis.keys()))

'''Создание второй структуры, идентичной первой, удаление оттуда записи 
про Копатыча так чтобы оригинальная структура не поменялась '''
smesharikis_new = smesharikis.copy()
smesharikis_new.pop("Копатыч")
print(smesharikis)
print(smesharikis_new)
```

```python
''' Задача 1 (3 балла)
Напишите простую программу для составления закодированных сообщений. Программа должна принимать на вход строку,
которую нужно закодировать. После этого она должна взять все нечётные элементы этой строки и заменить их таким
образом, чтобы первый нечётный элемент стал последним, второй нечётный элемент стал предпоследним нечётным элементом.
И так далее. Подробнее смотри в примерах:

>>> cat
>>> tac

>>> Word
>>> roWd

>>> Hello world
>>> derlw oollH
То есть все чётные буквы остаются на месте, а нечётные обращаются вспять.
'''

s = input()
#получаем символы на нечётных позициях
s_odd = s[1::2]
#получаем символы на чётных позициях в обратном порядке
s_even = s[::2][::-1]
result = "".join(sum(zip(s_even, s_odd), ())) + s[0] * (len(s) % 2)
print(result)
```
